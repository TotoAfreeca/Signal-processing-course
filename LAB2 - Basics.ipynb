## Zadanie 2

```python
def generate_signal(A, f, phi,fp ):
    tp= 1/fp
    dur = 0.008
    t = np.arange(0,dur+tp, tp)
    y = A*np.sin(2*np.pi*f*t+phi)
    return y

def generate_signal_longer(A, f, phi,fs ): #Przydatne do sprawdzania dźwięku
    tp= 1/fs
    dur = 1
    t = np.arange(0,dur+tp, tp)
    y = A*np.sin(2*np.pi*f*t+phi)
    return y

fp = 44100 #czestotliwosc próbkowania
tp = 1/fp #okres
dur = 0.008 #długośc trwania sygnału
t = np.arange(0,dur+tp, tp)
```

```python
#Różne amplitudy
plt.figure(2,figsize=(10,10))
plt.subplots_adjust(top=1.0,hspace=0.4)

plt.subplot(311)
y1 = generate_signal(1,440,0,44100)
plt.plot(t*1000,y1)
plt.xlabel('Czas [ms]')
plt.ylabel('Amplituda')
plt.grid()
plt.title("A = 1, f = 440, $\phi$ = 0 rad, fp = 44100")
plt.axis([0,8,-5,5])
plt.yticks([-5,0,5])

plt.subplot(312)
y2 = generate_signal(2,440,0,44100)
plt.plot(t*1000,y2)
plt.xlabel('Czas [ms]')
plt.ylabel('Amplituda')
plt.grid()
plt.title("A = 2, f = 440, $\phi$ = 0 rad, fp = 44100")
plt.axis([0,8,-5,5])
plt.yticks([-5,0,5])

plt.subplot(313)
y3 = generate_signal(5,440,0,44100)
plt.plot(t*1000,y3)
plt.xlabel('Czas [ms]')
plt.ylabel('Amplituda')
plt.grid()
plt.title("A = 5, f = 440, $\phi$ = 0 rad, fp = 44100")
plt.axis([0,8,-5,5])
plt.yticks([-5,0,5])

#testy dźwięku
print("Test 1")
ipd.display(ipd.Audio(generate_signal_longer(0.005,440,0,44100), rate = 44100))
print("Test 2")
ipd.display(ipd.Audio(generate_signal_longer(5,440,0,44100), rate = 44100))
print("Test 3")
ipd.display(ipd.Audio(generate_signal_longer(100,440,0,44100), rate = 44100))
```

```python
#Różne częstotliwości

plt.figure(3,figsize=(10,10))
plt.subplots_adjust(top=1.0,hspace=0.4)
plt.subplot(311)
y4 = generate_signal(1,1000,0,44100)
plt.yticks([-1,-0.5,0,0.5,1])
plt.plot(t*1000,y4)
plt.xlabel('Czas [ms]')
plt.ylabel('Amplituda')
plt.grid()
plt.title("A = 1, f = 1000, $\phi$ = 0 rad, fp = 44100")
plt.axis([0,8,-1,1])


plt.subplot(312)
y5 = generate_signal(1,2000,0,44100)
plt.plot(t*1000,y5)
plt.yticks([-1,-0.5,0,0.5,1])
plt.xlabel('Czas [ms]')
plt.ylabel('Amplituda')
plt.grid()
plt.title("A = 1, f = 2000, $\phi$ = 0, fp = 44100")
plt.axis([0,8,-1,1])


plt.subplot(313)
y6 = generate_signal(1,5000,0,44100)
plt.plot(t*1000,y6)
plt.yticks([-1,-0.5,0,0.5,1])
plt.xlabel('Czas [ms]')
plt.ylabel('Amplituda')
plt.grid()
plt.title("A = 1, f = 5000, $\phi$ = 0, fp = 44100")
plt.axis([0,8,-1,1])

print("Test 1")
ipd.display(ipd.Audio(generate_signal_longer(1,100,0,44100), rate = 44100))
print("Test 2")
ipd.display(ipd.Audio(generate_signal_longer(1,440,0,44100), rate = 44100))
print("Test 3")
ipd.display(ipd.Audio(generate_signal_longer(1,1200,0,44100), rate = 44100))
```
