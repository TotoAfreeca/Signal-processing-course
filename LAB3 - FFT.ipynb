```python
def generate_signal(A, N,phi ):
    n = np.arange(0,N,1)
    y = A*np.cos((1/A)*2*np.pi*n/N+phi)
    
    return n,y
x, y1 = generate_signal(1,32,np.pi/4)
x, y2 = generate_signal(0.5,32,0)
x, y3 = generate_signal(0.25,32,np.pi/2)
y4 = y1 + y2 + y3
plt.plot(x,y1,label='y1[n] = $cos(2\pi n/N+\pi /4)$')
plt.plot(x,y2, label='y2[n] = $0.5cos(4\pi n/N)$')
plt.plot(x,y3, label='y3[n] = $0.25cos(8\pi n/N+\pi /2)$')
plt.grid()
plt.ylabel('Wartość')
plt.xlabel('Numer Próbki')
plt.title('Wykresy funkcji cosinusowych')
plt.legend();
```

### FFT dla sinusa

```python

# funkcja dftmatrix - źródło
# http://python-for-signal-processing.blogspot.com/2013/02/discrete-fourier-transform.html
# zostało zmienione tylko dzielenie w ostatniej linijce z pierwiastka na liczbe całkowitą

def dftmatrix(Nfft=32,N=None):
    'construct DFT matrix'
    k= np.arange(Nfft)
    if N is None: N = Nfft
    n = np.arange(N)
    U = np.matrix(np.exp(1j* 2*np.pi/Nfft *k*n[:,None])) # use numpy broadcasting to create matrix
    return U/32

# Transformata
def dystrfo(y):
    n=y.size
    trans = y*dftmatrix(n,32)
    return trans


# Odwrócona Transformata
def idystrfo(y):
    n = y.size
    out = y*np.conj(dftmatrix(n,32))
    return out

x = np.arange(32)

y = np.sin(2*np.pi*x/32)
Y =  2*np.fft.fft(y)/32

Yw = 2*dystrfo(y)
Yw = np.squeeze(np.asarray(Yw))

plt.figure(1,figsize=(15,10))
plt.subplots_adjust(top=1.0,hspace=0.5)

plt.subplot(421)
plt.stem(x, np.real(y))
plt.title('Re(y)')
plt.xlabel('Numer próbki')
plt.ylabel('Amplituda')
plt.yticks([-1,-0.5,0,0.5,1])
plt.grid()

plt.subplot(422)
plt.stem(x,np.imag(y))
plt.title('Im(y)')
plt.xlabel('Numer próbki')
plt.ylabel('Amplituda')
plt.yticks([-1,-0.5,0,0.5,1])
plt.grid()

plt.subplot(423)
plt.stem(x, np.abs(Y))
plt.title('Moduł widma FFT')
plt.xlabel('Numer Pasma Częstotliwościowego')
plt.ylabel('Magnituda')
plt.grid()

plt.subplot(424)
plt.stem(x, np.abs(Yw))
plt.title('Moduł Widma - Funkcja Własna')
plt.xlabel('Numer Pasma Częstotliwościowego')
plt.ylabel('Magnituda')
plt.grid()


plt.subplot(425)
plt.stem(x, np.angle(Y)/np.pi)
plt.title('Faza widma FFT')
plt.xlabel('Numer Pasma Częstotliwościowego')
plt.ylabel('Faza [pi x rad]')
plt.grid()

plt.subplot(426)
plt.stem(x, np.angle(Yw)/np.pi)
plt.title('Faza Widma - Funkcja Własna')
plt.xlabel('Numer Pasma Częstotliwościowego')
plt.ylabel('Faza [pi x rad]')
plt.grid()


```
