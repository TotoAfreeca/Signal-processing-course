```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import signal
import IPython.display as ipd
from scipy.io import wavfile

dur = 5
fs = 8000
f = 1000

tp = 1/fs

t = np.arange(0,dur,tp)
sinus = np.sin(2*np.pi*f*t+0)


noise = np.random.normal(0,1,dur*fs)
zmienny = signal.chirp(t,0,5,1000)
fs, data = wavfile.read('voice.wav')
alfa = 0.999


T = len(t)
def obwiednia(y, T,a):
    obw = np.zeros(T)
    for n in range(1,T):
        obw[n] = a*obw[n-1]+(1-a)*(y[n])**2
    return obw

plt.figure(1,figsize=(15,10))
plt.subplots_adjust(top=1.0,hspace=0.4)

plt.subplot(421)
plt.plot(t,noise)
plt.title("Szum Gaussowski")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

gaussObwiednia = obwiednia(noise, T, 0.999)
plt.subplot(422)
plt.plot(t,gaussObwiednia)
plt.title("Szum Gaussowski - obwiednia mocy,  = 0.999")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')



plt.subplot(423)
plt.plot(t, sinus)
plt.title("Sinus o staej czstotliwoci")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

obwiedniaSin = obwiednia(sinus, T, 0.999)
plt.subplot(424)
plt.plot(t,obwiedniaSin)
plt.title("Obwiednia mocy sinus  = 0.999 ")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')


plt.subplot(425)
plt.plot(t, zmienny)
plt.title("Sinus o zmiennej czstotliwoci")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

obwZmienny = obwiednia(zmienny,T,0.999)
plt.subplot(426)
plt.plot(t, obwZmienny)
plt.title("Sinus o zmiennej czstotliwoci - Obwiednia mocy  = 0.999  ")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

plt.subplot(427)
plt.plot(t, data)
plt.title("Sygna mowy")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

obwiedniaVoice = obwiednia(data,T,0.999)
plt.subplot(428)
plt.plot(t, obwiedniaVoice)
plt.title("Sygna mowy - Obwiednia mocy,  = 0.999")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda');



```

## Wnioski 

Obwiednia mocy sygnau jest funkcj czasu, ktra zmienia si w wolniej ni sam sygna i jej
zadaniem jest wykrelenie krzywej ekstrem lokalnych funkcji sygnau w celu uatwienia analizy
przebiegu sygnau (okrela amplitud, ale nie czstotliwo czy faz). Pod wicze w celu
oszacowania mocy sygnau w czasie Px[n] zastosowane zostao urednianie rekursywne podane
wzorem:
Px[n] = Px[n-1] + (1-)x[n (1)

Warto mocy sygnau w czasie we wzorze (1) zaley od dwch czynnikw: od historii (poprzednie
elementy obwiedni) oraz od aktualnej wartoci mocy sygnau. Dobr parametru alfa, ktry okrela
priorytet co jest waniejsze jest kluczowym aspektem przy wykrelaniu obwiedni.

Najwikszym problemem periodogramu jest pogodzenie dwch zmiennych zalenych, wariancji
(mwicej o tym jak bardzo estymacja jest dobra) oraz obcienia (mwicego o tym czy wraz ze
zwikszaniem prbek estymacja zblia si do rzeczywistej wartoci).[4]

W przypadku wykorzystania okna prostoktnego otrzymujemy zbyt du warto wariancji, mamy
do czynienia z do silnymi wpywami z przeciekw widm. Celem jest redukcja (niwelacja jest
bowiem niemoliwa) przeciekw widm w taki sposb by odlego pomidzy gwnym prkiem a
pozostaymi bya jak najwiksza. W tym celu modyfikuje si periodogram splatajc go dodatkowo z
inn funkcj okna. Metoda ta jednak przy zmniejszaniu wariancji zwiksza obcienie. Innym
rozwizaniem jest metoda Welcha, ktra dodatkowo polega na nakadaniu si okien i urednianiu
zmodyfikowanych periodogramw w celu dodatkowej redukcji wariancji.[4]

Tworzc wasny spektrogram wykorzystujcy Short-time-fourier-transformate (funkcja STFT) moemy zauway drobne rnice w porwnaniu do tego "wbudowanego"- najbardziej widoczne s one na spektrogramach sygnau sinusoidalnego o staej czstotliwoci.
