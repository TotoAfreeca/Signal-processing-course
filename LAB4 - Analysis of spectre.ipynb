```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy import signal
import IPython.display as ipd
from scipy.io import wavfile

dur = 5
fs = 8000
f = 1000

tp = 1/fs

t = np.arange(0,dur,tp)
sinus = np.sin(2*np.pi*f*t+0)


noise = np.random.normal(0,1,dur*fs)
zmienny = signal.chirp(t,0,5,1000)
fs, data = wavfile.read('voice.wav')
alfa = 0.999


T = len(t)
def obwiednia(y, T,a):
    obw = np.zeros(T)
    for n in range(1,T):
        obw[n] = a*obw[n-1]+(1-a)*(y[n])**2
    return obw

plt.figure(1,figsize=(15,10))
plt.subplots_adjust(top=1.0,hspace=0.4)

plt.subplot(421)
plt.plot(t,noise)
plt.title("Szum Gaussowski")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

gaussObwiednia = obwiednia(noise, T, 0.999)
plt.subplot(422)
plt.plot(t,gaussObwiednia)
plt.title("Szum Gaussowski - obwiednia mocy, α = 0.999")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')



plt.subplot(423)
plt.plot(t, sinus)
plt.title("Sinus o stałej częstotliwości")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

obwiedniaSin = obwiednia(sinus, T, 0.999)
plt.subplot(424)
plt.plot(t,obwiedniaSin)
plt.title("Obwiednia mocy sinus α = 0.999 ")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')


plt.subplot(425)
plt.plot(t, zmienny)
plt.title("Sinus o zmiennej częstotliwości")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

obwZmienny = obwiednia(zmienny,T,0.999)
plt.subplot(426)
plt.plot(t, obwZmienny)
plt.title("Sinus o zmiennej częstotliwości - Obwiednia mocy α = 0.999  ")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

plt.subplot(427)
plt.plot(t, data)
plt.title("Sygnał mowy")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda')

obwiedniaVoice = obwiednia(data,T,0.999)
plt.subplot(428)
plt.plot(t, obwiedniaVoice)
plt.title("Sygnał mowy - Obwiednia mocy, α = 0.999")
plt.xlabel('Czas [s]')
plt.ylabel('Amplituda');



```

## Wnioski 

Obwiednia mocy sygnału jest funkcją czasu, która zmienia się w „wolniej” niż sam sygnał i jej
zadaniem jest wykreślenie krzywej ekstrem lokalnych funkcji sygnału w celu ułatwienia analizy
przebiegu sygnału (określa amplitudę, ale nie częstotliwość czy fazę). Pod ćwiczeń w celu
oszacowania mocy sygnału w czasie Px[n] zastosowane zostało uśrednianie rekursywne podane
wzorem:
Px[n] = αPx[n-1] + (1-α)x[n (1)

Wartość mocy sygnału w czasie we wzorze (1) zależy od dwóch czynników: od historii (poprzednie
elementy obwiedni) oraz od aktualnej wartości mocy sygnału. Dobór parametru alfa, który określa
priorytet co jest ważniejsze jest kluczowym aspektem przy wykreślaniu obwiedni.

Największym problemem periodogramu jest pogodzenie dwóch zmiennych zależnych, wariancji
(mówiącej o tym jak bardzo estymacja jest dobra) oraz obciążenia (mówiącego o tym czy wraz ze
zwiększaniem próbek estymacja zbliża się do rzeczywistej wartości).[4]

W przypadku wykorzystania okna prostokątnego otrzymujemy zbyt dużą wartość wariancji, mamy
do czynienia z dość silnymi wpływami z przecieków widm. Celem jest redukcja (niwelacja jest
bowiem niemożliwa) przecieków widm w taki sposób by odległość pomiędzy głównym prążkiem a
pozostałymi była jak największa. W tym celu modyfikuje się periodogram splatając go dodatkowo z
inną funkcją okna. Metoda ta jednak przy zmniejszaniu wariancji zwiększa obciążenie. Innym
rozwiązaniem jest metoda Welcha, która dodatkowo polega na nakładaniu się okien i uśrednianiu
zmodyfikowanych periodogramów w celu dodatkowej redukcji wariancji.[4]

Tworząc własny spektrogram wykorzystujący Short-time-fourier-transformate (funkcja STFT) możemy zauważyć drobne różnice w porównaniu do tego "wbudowanego"- najbardziej widoczne są one na spektrogramach sygnału sinusoidalnego o stałej częstotliwości.
